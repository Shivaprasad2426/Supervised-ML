# Explain Bagging and Boosting methods. How is it different from each other.

## Bagging (Bootstrap Aggregating)
## Bagging is an ensemble method that combines multiple "weak learners" (typically decision trees) to create a more robust and accurate "strong learner."  Here's how it works:

## 1. Bootstrap Sampling:  Multiple subsets of the training data are created by randomly sampling the original dataset *with replacement*.  This means that some data points may appear multiple times in a single subset, while others may be left out.

## 2. Model Training: A separate model (e.g., a decision tree) is trained on each of these bootstrapped subsets. Because the datasets are slightly different, the individual models will also be slightly different.



## Boosting
### Boosting is another ensemble method that combines multiple weak learners but in a sequential manner. The key difference from bagging is that boosting trains models sequentially, and each subsequent model attempts to correct the errors made by its predecessors.

## 1. Sequential Training:
### The first model is trained on the entire dataset.  The second model is trained on a modified version of the dataset where the data points that the first model misclassified are given higher weights.

## 2. Weighted Data Points:
### The misclassified data points are given more importance in the subsequent models. Each model focuses on the data points that previous models struggled with.

## 3. Combining Predictions:
### Similar to bagging, the final prediction is an aggregation of all model predictions.  However, the contributions of different models are weighted based on their performance.  Boosting tends to reduce bias.

## Differences between Bagging and Boosting:

## - Data Sampling:
### Bagging uses bootstrap sampling (random sampling with replacement), while boosting assigns weights to data points and focuses on misclassified samples.
## - Model Training: Bagging trains models in parallel, whereas boosting trains models sequentially.
## - Error Correction:
### Bagging reduces variance by averaging predictions of multiple models, while boosting reduces bias by iteratively focusing on the mistakes of previous models.
## - Model Weights:
### Bagging treats all models equally during aggregation, while boosting assigns weights to models based on their performance.
## - Overfitting:
### Bagging is less prone to overfitting than boosting. Boosting can overfit if not carefully tuned (e.g., number of estimators).
